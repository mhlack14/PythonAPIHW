```python
#Create new DataFrame fitting weather criteriaÂ¶
#Narrow down the cities to fit weather conditions.
#Drop any rows will null values.
```

```python
# Narrow down cities that fit criteria i.e. Max temp less than 80 but greater than 70 with wind speeds less than 10mph
#...and drop any results with null values. Display results

CriteriaCities = cities.loc[(cities["Max Temp"] < 80) & 
                                    (cities["Max Temp"] > 70) \
                                    & (cities["Wind Speed"]<10) \
                                    & (cities["Cloudiness"]==0)].dropna()
#check to make sure code works properly, visually. 50 columns is enough to see per the .head function. 
CriteriaCities.head(50)
```

```python
#Hotel MapÂ¶
#store into variable named hotel_df.
#Add a "Hotel Name" column to the DataFrame.
#Set parameters to search for hotels with 5000 meters.
#Hit the Google Places API for each city's coordinates.
#Store the first Hotel result into the DataFrame.
#Plot markers on top of the heatmap.
```

```python
#Store into variable named hotel_df
hotel_df = CriteriaCities[["City", "Country", "Lat", "Lng"]].copy()

hotel_df
```

```python
#Hotel Name
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
#Parameters to be set
parameters = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key
}
#check to see paramters code works
parameters
```

```python
#Hit the Google Places API for each city's coordinates.
 
for index, row in hotel_df.iterrows():
    # get lat, lng from df
    lat = row["Lat"]
    lng = row["Lng"]
    parameters["location"] = f"{lat},{lng}"
    #Reference the google data
    URL="https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    URLrequest = requests.get(URL, params=parameters)
    #Jsonify
    URLrequest=URLrequest.json()
    
    try:
        hotel_df.loc[index, "Hotel Name"] = URLrequest["results"][0]["name"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer on top of heat map
marker = gmaps.marker_layer(locations, info_box_content=hotel_info)
figure.add_layer(marker)

# Display figure
print(figure)
```
